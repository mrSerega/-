```python
import numpy as np
import cv2 
from matplotlib import pyplot as plt
import sys
import time
```

##  

```python
def SWAHE(
    img, #  
    r, #   SWAHE
    show_progress # ,    
):
    clahe = cv2.createCLAHE( clipLimit=2.0)

    height = img.shape[0]
    width = img.shape[1]

    full = (height-2*r)*(width-2*r)

    if show_progress: print ('start')  #   (1)

    count = 0
    step_to_show = 1
    d = r

    new_pic = np.zeros((height,width))


    for h in range(r,height-r, int(r/d)):
        for w in range(r,width-r, int(r/d)):

            count += 1

            if show_progress:
                #    (2)
                step = (h-r)*(width-2*r)+(w-r) 
                sys.stdout.flush()
                sys.stdout.write('{} / {} % ({}/{})\r'.format(int(step/(full-1) * 100),100, step, full))
                #    (2)

            left_up_h = h-r
            left_up_w = w-r
            right_down_h = h+r
            right_down_w = w+r

            if left_up_h < 0: left_up_h = 0
            if left_up_w < 0: left_up_w = 0
            if right_down_h < 0: right_down_h = 0
            if right_down_w < 0: right_down_w = 0

            if left_up_h > height-1: left_up_h = height-1
            if left_up_w > width-1: left_up_w = width-1
            if right_down_h > height-1: right_down_h = height-1
            if right_down_w > width-1: right_down_w = width-1

            sl = img[left_up_h:right_down_h,left_up_w:right_down_w]
            sl = clahe.apply(sl)
            new_pic[left_up_h:right_down_h,left_up_w:right_down_w] = sl[:]
    if show_progress: print ('done')
    return new_pic

    
```

 #  

##  

```python
image_path = 'clahe_1.jpg'   #    
r = 100   #   SWAHE
show_progress = True # ,    
output_mode = 0 #       (0 --  , 1 --   )
result_name = 'newpic.jpg' #   
```

## 

```python
img = np.array(cv2.imread(image_path,0)) #    (   )
new_pic = SWAHE(img, r, show_progress) #  
if output_mode == 0:
    plt.imshow(new_pic, cmap='gray')          #    
elif output_mode == 1:
    cv2.imwrite(result_name ,new_pic)          #    
```
